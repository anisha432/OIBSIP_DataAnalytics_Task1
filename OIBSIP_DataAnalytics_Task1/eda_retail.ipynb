```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

# 1 Data Loading and Cleaning: Load the retail sales dataset

```python
df = pd.read_csv("retail_sales_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

# 2 Descriptive Statistics: Calculate basic statistics (mean, median, mode, standard deviation)

```python
df.describe()
```

```python
df.columns
```

# 3 Time Series Analysis: Analyze sales trends over time using time series technique

```python
df['Date'] = pd.to_datetime(df['Date'])
df.groupby('Date')['Total Amount'].sum().plot()
plt.title("Sales Trend Over time")
plt.show()
```

# 4 Customer and Product Analysis: Analyze customer demograohics and purchasing behavior

```python
df['Product Category'].value_counts()
```

```python
df.groupby('Product Category')['Total Amount'].sum()
```

```python
sns.histplot(df['Total Amount'], bins = 10)
plt.title("Product category Count")
plt.show()
```

# 5 Visualization: Present insight through bar charts, line plots, and heatmaps

```python
df['Product Category'].value_counts().plot(kind ='bar')
plt.title("Product Category Count")
plt.show()
```
